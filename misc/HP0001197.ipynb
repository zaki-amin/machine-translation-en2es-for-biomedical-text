```python
import pandas as pd

filename = "HP0001197.csv"
df = pd.read_csv(filename)
```

```python
def select_category(category: str):
    sub_df = df[df[category] == 'X']
    sub_df = sub_df.drop(["Mejor", "Igual", "Peor", "Comentarios Raúl", "Comentarios H12O"], axis=1)
    return sub_df


def summarise_category(category: str):
    subset = select_category(category)
    print(f"{category} - {len(subset)}")
    return subset
```
